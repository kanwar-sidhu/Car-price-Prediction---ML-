```python
df.drop_duplicates(inplace=True)
```

Check Unique Value

```python
df.nunique()
```

```python
df.describe()
```

```python
#In {owner} coloumn remove owner from first owner
df['owner'] = df['owner'].str.replace(' Owner', '')
```

```python
df
```

```python
#make all string values in data set (lower case)
for column in df.select_dtypes(include=[object]):
    df[column] = df[column].str.lower()
```

```python
df
```

```python
#check unique category in each coloumn
categorical_column = ['name','year','fuel','seller_type','transmission','owner',]

for col in categorical_column:
    print(f"Category in {col} is : {df[col].unique()}")
```

```python
frequency = ['name','year','fuel','seller_type','transmission','owner']

for fre in frequency:
    print(f"frequency in {fre} is : {df[fre].value_counts()}")
```

```python
# Identify names with counts greater than 10
frequent_names = df['name'].value_counts()
frequent_names = frequent_names[frequent_names > 10].index.tolist()

# Filter the DataFrame based on these names
df = df[df['name'].isin(frequent_names)]
df
```

```python
# Since the DataFrame is already filtered by 'name', recalculate 'year' value counts
frequent_years = df['year'].value_counts()
frequent_years = frequent_years[frequent_years > 5].index.tolist()

# Further filter the DataFrame based on these years
df = df[df['year'].isin(frequent_years)]
df
```

```python
# Since the DataFrame is now filtered by both 'name' and 'year', recalculate 'fuel' value counts
frequent_fuels = df['fuel'].value_counts()
frequent_fuels = frequent_fuels[frequent_fuels > 20].index.tolist()

# Further filter the DataFrame based on these fuel types
df = df[df['fuel'].isin(frequent_fuels)]
df
```

```python
# Since the DataFrame is now filtered by both 'name' and 'year', recalculate 'seller_type' value counts
frequent_seller = df['seller_type'].value_counts()
frequent_seller = frequent_seller[frequent_seller > 35].index.tolist()

# Further filter the DataFrame based on these seller types
df = df[df['seller_type'].isin(frequent_seller)]
df
```

```python
# Since the DataFrame is now filtered by both 'name' and 'year', recalculate 'owner' value counts
frequent_owner = df['owner'].value_counts()
frequent_owner = frequent_owner[frequent_owner > 20].index.tolist()

# Further filter the DataFrame based on these owner types
df = df[df['owner'].isin(frequent_owner)]
df
```
